#### 데이터프레임 변수의 종류

```python
# total_bill 의 탑10 내림차순
total10 = tip.sort_values('total_bill', ascending=False).head(10)

# total_bill 인덱스 새로 채우고, 기존 인덱스 삭제
total10.reset_index(drop=True, inplace=True)
print(total10)
```

```python
import matplotlib.pyplot as plt

plt.bar(total10.index, total10['total_bill'])
plt.title('My first bar chart', fontsize=20)
```
