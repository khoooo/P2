```python
import pandas as pd

titanic = pd.read_csv('https://bit.ly/TitanicFile')
titanic
```

#### 1-1. passengerld, Name, Ticket, Cabin 열을 한 번에 삭제하시오.

```python
x_cols = ['PassengerId', 'Name', 'Ticket', 'Cabin']
titanic.drop(x_cols, axis=1, inplace=True)
titanic
```

#### 1-2. sex 열 이름을 Male로 변경

```python
cols = {'Sex':'Male'}
titanic.rename(columns=cols, inplace=True)
titanic
```

#### 1-3. Male 열 값을 male=0, female=1 로 바꾸시오.

```python
titanic['Male'] = titanic['Male'].map({'male':0, 'female':1})
titanic
```

#### 1-4. SibSp 열과 parch 열의 값을 더한 결과를 Family 열에 추가

```python
titanic['Family'] = titanic['SibSp']+titanic['Parch']
titanic
```

#### 1-5. SibSp, Parch 열을 삭제하시오.

```python
titanic.drop(['SibSp', 'Parch'], axis=1, inplace=True)
titanic
```

#### 2-1. 기초통계량에서 Age 열의 사분위수 조회하여 q1,q2,q3에 저장

```python
q1 = titanic.describe().loc['25%','Age']
q2 = titanic.describe().loc['50%','Age']
q3 = titanic.describe().loc['75%','Age']
print(q1)
print(q2)
print(q3)
```

#### 2-2. 아래 범위를 참고하여 AgeGrp 열을 추가
 - q1 이하 -> 'a'
 - q1 초과 q2 이하 -> 'b'
 - q2 초과 q3 이하 -> 'c'
 - q3 초과 -> 'd'

```python
import numpy as np

a = [-np.inf,q1,q2,q3,np.inf]
b = ['a','b','c','d']

titanic['AgeGrp'] = pd.cut(titanic['Age'], bins=a, labels=b)
titanic
```

#### 2-3. AgeGrp 별 생존율 집계하여 tmp 데이터프레임에 선언

```python
tmp = titanic.groupby('AgeGrp')[['Survived']].mean()
tmp.reset_index(drop=False, inplace=True)
tmp
```

#### 2-4. tmp를 막대그래프로 시각화하시오.

```python
import matplotlib.pyplot as plt

plt.bar(tmp['AgeGrp'],tmp['Survived'])
plt.show()
```
