```python
import pandas as pd
pop1 = pd.read_csv('https://bit.ly/PopH01File', index_col='year')
pop2 = pd.read_csv('https://bit.ly/PopH02File', index_col='year')
pop3 = pd.read_csv('https://bit.ly/PopH03File', index_col='year')
pop2
```

#### concat
 - outer
 - inner 
 - 여러개 한 번에 가능.
 - 인덱스를 지정해줘야 함.

```python
# 81부터 시작
pop = pd.concat([pop1,pop2,pop3], join='outer', axis=1)
pop
```

```python
#85 부터 시작
pop = pd.concat([pop1,pop2,pop3], join='inner', axis=1)
pop
```

#### merge
 - 키를 위주로 합병
 - 동시에 두개밖에 못함.
 - 기존 인덱스 필요 없음.

```python
pop1 = pd.read_csv('https://bit.ly/PopH01File')
pop2 = pd.read_csv('https://bit.ly/PopH02File')
pop3 = pd.read_csv('https://bit.ly/PopH03File')
```

```python
pop = pd.merge(pop1,pop2, on='year', how='inner')
pop = pd.merge(pop,pop3, on='year', how='inner')
pop
```
