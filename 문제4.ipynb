#### gre 평균을 gre_mean으로 선언

```python
gre_mean = graduate['gre'].mean()
gre_mean
```

#### 요구사항을 작성하시오.
- gre 가 800 이상인 행을 g1 데이터프레임으로 선언
- g1의 gpa 가 4.0 이상인 행을 g2 데이터프레임으로 선언
- g2 인덱스 초기화

```python
t1 = graduate['gre'] >= 800
t2 = graduate['gpa'] >= 4.0

g1 = graduate[t1]
g2 = graduate[t1 & t2]
g2.reset_index(drop=True, inplace=True)
g2
```

#### gre 가 평균 이상이면서 rank 가 4인 행을 g3 으로 선언

```python
t3 = graduate['gre'] >= graduate['gre'].mean()
t4 = graduate['rank'] == 4
g3 = graduate[t3 & t4]
g3
```
