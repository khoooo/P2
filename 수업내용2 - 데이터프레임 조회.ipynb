```python
# 남자고객의 tip, day, time 만 추출하여 tip2 에 선언
tip2 = tip.loc[tip['sex'] == 'Male', ['tip', 'day', 'time']]
tip2
```

```python
# isin - day 에서 Sun or Sat
tip.loc[tip['day'].isin(['Sun', 'Sat'])]
```

```python
# between - size 의 1,3 사이
tip.loc[tip['size'].between(1,3)]
```

```python
c1 = tip['size'] == 1
c2 = tip['day'].isin(['Fri','Thur'])

tip.loc[ c1 & c2, ['total_bill'] ]
```

#### 조회 연습
- 요일 ='토', '일'
- 'total_bill' 내림차순
- 인덱스 리셋
- top10

```python
tmp = tip.loc[tip['day'].isin(['Sat', 'Sun']), :]
temp = tmp.sort_values('total_bill', ascending=False)
temp.reset_index(drop=True, inplace=True)
top10 = temp.head(10)
top10
```

#### 집계
- groupby 로 'day'를 묶음.  
- 
- as_index -> 인덱스 풀어줌.

```python
tmp = tip.groupby('day', as_index=False)['total_bill'].sum()
tmp.sort_values('total_bill', ascending=False, inplace=True)
tmp
```

```python
import matplotlib.pyplot as plt

# plt.bar(tmp['day'], tmp['total_bill'])
plt.plot(tip['total_bill'])
plt.plot(tip['tip'])

plt.legend(['total_bill', 'tip'])
plt.ylabel('($)')
plt.title('total_bill_day', fontsize=15)

plt.show()
```
