```python
import pandas as pd

path = 'https://bit.ly/Tips2File'
tip = pd.read_csv(path)

tip
```

#### 열 이름 변경
- 일부 열 
- 모든 열

```python
#일부 열

cols = {'total_bill_amount' : 'total_bill', 
       'male_female' : 'sex',
       'week_name' : 'day'}
tip.rename(columns=cols, inplace=True)
tip
```

```python
#모든 열

tip.columns = ['total_bill', 'tip', 'sex', 
               'smoker', 'day', 'time','size']
tip
```

----

#### 열 추가

```python
tip['total_amt'] = tip['total_bill'] + tip['tip']
tip
```

#### 원하는 열에 추가

```python
tip.insert(1, 'tip_tb', tip['total_bill']/tip['size'])
tip
```

```python
tip['div_tb'] = tip['total_bill']/tip['size']
tip = tip[['total_bill', 'tip_tb', 'div_tb', 
          'sex', 'smoker', 
          'day', 'time', 'size', 'total_amt']]

tip
```

----

####  ### hoilday 열 추가하기 

```python
tip['hoilday'] = 0
tip.loc[tip['day'].isin(['Sat', 'Sun']), 'hoilday'] = 1

tip
```

```python
tip['hoilday'].value_counts()
```

#### 열 삭제

```python
x_cols = ['tip_tb', 'div_tb', 'hoilday']
tip.drop(x_cols, axis=1, inplace=True)
tip
```

#### 추가 기능
 - map : 범주형 데이터로 바꾸기
 
 
 - cut : 범위별 등급 지정
  - 지정하지 않는 단계
  - 지정하는 단계

```python
# sex 열의 데이터 개수에 맞게 지정해줘야함.

tip['sex'] = tip['sex'].map({'Male':1, 'Female':0})
tip
```

```python
tip['tip_grp'] = pd.cut(tip['tip'], 5, labels=list('edcba'))
tip
```

```python
tip['tip_grp'].value_counts()
```

```python
tip.describe()
```

```python
import numpy as np

q1 = tip.describe().loc ['25%', 'total_bill']
q2 = tip.describe().loc ['50%', 'total_bill']
q3 = tip.describe().loc ['75%', 'total_bill']

bin = [-np.inf,q1,q2,q3,np.inf]
label = ['bronze', 'silver', 'gold', 'platinum']

tip['total_grp'] = pd.cut(tip['total_bill'], bins=bin, labels=label)
print(bin)
print(tip)
```

```python
tip['total_grp'].value_counts()
```

#### 결측치

```python
import pandas as pd
air = pd.read_csv('https://bit.ly/AirQFile')
air
```

#### 결측치 확인

```python
air.info()
```

```python
#air.isna()
air.isnull()
```

```python
#널값 개수
air.isna().sum()
```

----

#### 결측치 제거
 - 행 제거

```python
air.dropna(axis=0, inplace=True)
air
```

#### 결측치 대체
 - 평균값
 - 중앙값
 - 바로 앞, 뒤 데이터 : ffill, bfill
 - 선형보간법

```python
oz_mean = air['Ozone'].mean()
air['Ozone'].fillna(oz_mean, inplace=True)
air.info()
```

```python
sol_mid = air['Solar.R'].median()
air['Solar.R'].fillna(sol_mid, inplace=True)
air.info()
```

```python
air['Solar.R'].interpolate(method='linear', inplace=True)
air
```
