```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

titanic = pd.read_csv('https://bit.ly/TitanicFile')
titanic
```

```python
titanic.info()
```

#### 1-1. Age 열의 결측치를 평균으로 채우시오.

```python
age_mean = titanic['Age'].mean()

titanic['Age'].fillna(age_mean, inplace=True)
titanic.info()
```

#### 1-2. Embarked 결측치를 'S' 로 채우시오.

```python
titanic['Embarked'].fillna('S', inplace=True)
titanic.info()
```

#### 1-3. passengerld, Name, Ticket, Cabin 열을 한 번에 삭제하시오.

```python
titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
titanic.info()
```

```python
titanic
```

#### 2-1. Pclass 별 개수를 tmp 에 선언

```python
tmp = titanic.groupby('Pclass', as_index=False)[['Survived']].count()
tmp
```

#### 2-2. 막대그래프를 그리시오.

```python
plt.bar(tmp['Pclass'], tmp['Survived'])
plt.show()
```

#### 3. Pclass, sex, Emberked 를 한 번에 가변수화하시오.
 - 다중공선성의 문제를 범주값 개수보다 1개 작은 개수로 하시오.

```python
cols = ['Pclass', 'Sex', 'Embarked']
titanic = pd.get_dummies(titanic, columns=cols, drop_first=True)
titanic
```
