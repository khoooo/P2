#### 5. 요구사항에 따라 확인하시오.
- tip 데이터프레임 total_bill이 45.0 이상, tip이 6.0 이상인 행을 good으로 선언
- good 데이터프레임 인덱스를 초기화하시오.
- good 데이터프레임을 확인하시오.

```python
t1 = tip['total_bill'] >= 45
t2 = tip['tip'] >= 6
good = tip[t1 & t2]
good.reset_index(drop=True, inplace=True)
good
```

```python
t1 = tip['total_bill'] >= 45.0
t2 = tip['tip'] >= 6.0
good = tip[t1 & t2]
good.reset_index(drop=True, inplace=True)
good
```
