#### 7. 요구사항에 따라 작성하시오.
- total_bill 열 기준으로 내림차순 정렬하여 상위 10개를 tip_top10 으로 선언
- tip_top10 데이터프레임 인덱스를 초기화
- tip_top10 데이터프레임 확인

```python
tip_top10 = tip.sort_values('total_bill', ascending=False).head(10)
tip_top10.reset_index(drop=True, inplace=True)
tip_top10
```

```python
tip_top10 = tip.sort_values('total_bill').head(10)
tip_top10.reset_index(drop=True, inplace=True)
tip_top10
```
